## Operaciones más avanzadas

## GROUP BY

### Tarea: Crear una tabla que nos muestre el TOP 10 de valor promedio de mercado de jugadores, agrupado por club actual y ordenado de mayor promedio a menor promedio.

```python
# Configurar pandas para mostrar números en formato completo
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
q = """ 
SELECT
current_club_name AS nombre_del_club,
AVG(highest_market_value_in_eur) AS promedio_valor
FROM jugadores
GROUP BY 1
ORDER BY promedio_valor DESC
LIMIT 11
"""

top_10_clubes = pysqldf(q)
```

```python
top_10_clubes
```

## HAVING

### Tarea: Queremos la misma tabla pero teniendo en cuenta solo clubes con valor promedio menor a 15,000,000 EUR.
### En este caso aplicamos HAVING y no WHERE ya que el filtro lo aplicamos sobre el valor ya agrupado, cuando aplicamos WHERE en el primer ejercicio lo hacemos directamente sobre la fila cruda que estamos consultando.

```python
q = """ 
SELECT
current_club_name AS pais_residencia,
AVG(highest_market_value_in_eur) AS promedio_valor
FROM jugadores
GROUP BY 1
HAVING promedio_valor < 15000000
ORDER BY promedio_valor DESC
LIMIT 11
"""

pysqldf(q)
```

### Ahora lo que queremos es saber cómo se llama el estadio de cada uno de estos clubes. Para eso vamos a tener que hacer un JOIN entre la tabla que creamos y la tabla de clubes que aprece acá abajo y que importamos al principio del ejercicio.

### Para hacer el JOIN realizamos lo siguiente. Primero hay que tener en cuenta que existen diferentes tipos de JOIN como se muestra en la siguiente imagen. Para nuestro ejemplo vamos a usar un LEFT JOIN, que es uno de los más utilizados, o por lo menos el que más utilizo yo generalmente.

```python
# Especifica la ruta de la imagen local
ruta_imagen = 'Visual_SQL_JOINS_orig.jpg'

# Muestra la imagen en el notebook
Image(filename=ruta_imagen)
```

```python
q = """ 
SELECT 
a.*,
b.stadium_name
FROM top_10_clubes a
LEFT JOIN clubes b
ON a.nombre_del_club = b.name
"""

pysqldf(q)
```

### De esta forma podemos hacer infinitas transformaciones y generar DataFrames de la nada a partir de SQL, lo cuál vuelve el trabajo en Python muchísimo más fácil si estamos acostumbrados a usar SQL.

### Si queríamos hacer lo mismo en Pandas usando Python, se vería algo así. Primero calculamos la tabla del TOP 10:

```python
# Supongamos que tu DataFrame se llama jugadores
# Replace 'nombre_de_jugador', 'current_club_name', y 'highest_market_value_in_eur' con tus nombres de columna reales

# Paso 1: Agrupar por 'current_club_name' y calcular el promedio
promedio_por_club = jugadores.groupby('current_club_name')['highest_market_value_in_eur'].mean()

# Paso 2: Ordenar por el promedio descendente
resultado_ordenado = promedio_por_club.sort_values(ascending=False)

# Paso 3: Seleccionar las primeras 10 filas (top 10)
top_10_clubes = resultado_ordenado.head(10)

# Imprimir el resultado
top_10_clubes
```

```python
# Supongamos que top_10_clubes es el DataFrame que obtuviste anteriormente
# Supongamos que clubes es el DataFrame con la información de los clubes

# Paso 1: Realizar el join utilizando merge
resultado_final = top_10_clubes.reset_index().merge(clubes, left_on='current_club_name', right_on='name', how='left')

# Paso 2: Eliminar columnas innecesarias
resultado_final = resultado_final[['current_club_name', 'highest_market_value_in_eur', 'stadium_name']]

# Imprimir el resultado final
resultado_final
```
