### Manipulación y Filtrado de Datos

```python
# Seleccionar una columna específica para análisis
columna_seleccionada = 'release_year'
print(f"\nAnálisis de la columna '{columna_seleccionada}':")
print(df_netflix[columna_seleccionada].value_counts())
```

```python
# Filtrar shows con country = United States y release_year mayor a 2005
filtro_shows = (df_netflix['country'] == 'United States') & (df_netflix['release_year'] > 2005)

# Aplicar el filtro y crear un nuevo DataFrame con los resultados
df_netflix_filtrado = df_netflix[filtro_shows]

# Mostrar los resultados
print("Shows de Estados Unidos con fecha de estreno posterior al 2005:")
df_netflix_filtrado
```
